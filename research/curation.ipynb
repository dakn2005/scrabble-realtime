```python
import nltk
from nltk.corpus import stopwords


enStop = stopwords.words('english')
enStop
```

```python
def getLines(filename, fpath = None):
    lines = []
    lepath = '../server/utils/dictionaries' if not fpath else fpath

    with open(f'{lepath}/{filename}') as f:
        lines = f.readlines()
    
    lines = [str(l).replace('\n', '') for l in lines if len(str(l).replace('\n', '')) > 1]

    return lines

def writetLines(arr, filename, fpath = '.'):
    with open(f"{fpath}/{filename}", "a") as f:
        for wrd in arr:
            f.write(wrd+'\n')
```

```python
enArr = getLines('en.txt','resources')
swaArr = getLines('swahili.txt','resources')
swaStop = getLines('swa_stop.csv','resources')

swaStop = swaStop[1:]
# swaStop
```

```python
def curator(arr, comparator) -> list:
    out = set()

    for word in arr:
        word = word.replace("'", "")
        
        # if word not in out:
        if len(word) < 3:
            if word in comparator:
                out.add(word)
        else:
            out.add(word)

    return list(out)
```

```python
en_curated = curator(enArr, enStop)
swa_curated = curator(swaArr, swaStop)
en_curated.sort()
swa_curated.sort()
writetLines(en_curated, 'en_curated.txt')
writetLines(swa_curated, 'swa_curated.txt')
```
