```python
from collections import Counter
from matplotlib import pyplot as plt
```

```python
def getLines(filename):
    lines = []

    with open(f'../server/utils/dictionaries/{filename}.txt') as f:
        lines = f.readlines()
    
    lines = [str(l).replace('\n', '') for l in lines if len(str(l).replace('\n', '')) > 1]
    return lines

def getShengLines(filename):
    lines, out = [], []

    with open(f'../server/utils/dictionaries/{filename}.txt') as f:
        lines = f.readlines()
    
    for ll in lines:
        if len(ll) > 1:
            wrd = ll.replace('\n', '').split()[0]
            out.append(wrd)

    return out
```

```python
enArr = getLines('en')
shengArr = getShengLines('sheng')
swaArr = getLines('swahili')

print(enArr[:5], '\n', shengArr[:5], '\n',swaArr[:5])
# print(shengArr[:5])
```

### Get All Letters

```python
def isLetter(l):
    upperL = l.upper()
    return ord('A') <= ord(upperL) <= ord('Z')

def allLetters(arr):
    out = []

    for elem in arr:
        for letter in elem:
            upL = letter.upper()
            
            if isLetter(upL):
                out.append(upL)

    return out
```

```python
enLCounter = Counter(allLetters(enArr))
shengLCounter = Counter(allLetters(shengArr))
swaLCounter = Counter(allLetters(swaArr))

# print(swaLCounter)
```

### Hist

```python
# https://www.tutorialspoint.com/how-to-plot-collections-counter-histogram-using-matplotlib
def histogram(lCounter: Counter, title):
    # https://www.freecodecamp.org/news/python-sort-dictionary-by-key/
    # lKeySorted = dict(sorted(lCounter.items()))
    lValueSorted = dict(lCounter.most_common(27))

    plt.rcParams["figure.figsize"] = [7.50, 3.50]
    plt.rcParams["figure.autolayout"] = True

    # plt.bar(lCounter.keys(), lCounter.values())
    # plt.bar(lKeySorted.keys(), lKeySorted.values())
    plt.bar(lValueSorted.keys(), lValueSorted.values())
    plt.title(title)
    plt.show()
```

```python
histogram(enLCounter, 'English Distrn')
```

```python
histogram(swaLCounter, 'Swahili Distrn')
```

```python
histogram(shengLCounter, 'Sheng Distrn')
```
